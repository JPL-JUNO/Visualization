## Introduction to Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')

x = np.arange(0, 10.5, .2)
y = np.sin(x)
z = np.cos(x)
plt.plot(x, y)
plt.plot(x, z)
plt.show()
```

## The definition of figure, plots, subplots, axes, ticks, and legends

```python
for index, fun in enumerate(plt.get_plot_commands(), 1):
    if index % 10 == 0:
        print(fun)
    print(fun, end=", ")
```
