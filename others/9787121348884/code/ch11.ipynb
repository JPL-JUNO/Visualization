```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False
plt.axes([0.1, 0.1, 0.8, 0.8], frameon=True, facecolor="y", aspect="equal")
plt.plot(2 + np.arange(3), [0, 1, 0])
plt.text(2.25, 0.8, "FONT")
plt.title("Line Chart")
plt.show()
```

```python
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False
plt.rcParams["lines.linewidth"] = 8.0
plt.rcParams["lines.linestyle"] = "--"

# font properties in change
plt.rcParams["font.style"] = "normal"
plt.rcParams["font.variant"] = "small-caps"
plt.rcParams["font.weight"] = "black"
plt.rcParams["font.size"] = 15.0

# text properties in change
plt.rcParams["text.color"] = "blue"
plt.axes([0.1, 0.1, 0.8, 0.8], frameon=True, facecolor="y", aspect="equal")
plt.plot(2 + np.arange(3), [0, 1, 0])
plt.text(2.25, 0.8, "FONT")
plt.title("Line Chart")
plt.show()
```

```python
plt.axes([0.1, 0.1, 0.8, 0.8], frameon=True, facecolor="y", aspect="equal")
# line properties in change
plt.plot(2 + np.arange(3), [0, 1, 0], linewidth=8.0, linestyle="--")
plt.title(
    "Line Chart",
    color="red",
    style="normal",
    variant="small-caps",
    weight="black",
    size=18,
)
plt.text(
    2.25,
    0.8,
    "FONT",
    color="blue",
    fontdict={
        "style": "normal",
        "variant": "small-caps",
        "weight": "black",
        "size": 28,
    },
)
plt.title("Line Chart")
plt.show()
```
