```python
import matplotlib.pyplot as plt
import numpy as np
```

## 使用颜色参数和颜色映射表

```python
barSlices = 12
theta = np.linspace(0.0, 2 * np.pi, barSlices, endpoint=False)
radii = 30 * np.random.rand(barSlices)
width = 2 * np.pi / barSlices
colors = np.array(
    [
        "c",
        "m",
        "y",
        "b",
        "#C67171",
        "#C1CDCD",
        "#FFEC8B",
        "#A0522D",
        "red",
        "burlywood",
        "chartreuse",
        "green",
    ]
)
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
bars = ax.bar(theta, radii, width=width, color=colors, bottom=0.0)
plt.show()
```

```python
hexHtml = [
    "#d73027",
    "#f46d43",
    "#fdae61",
    "#fee090",
    "#ffffbf",
    "#e0f3f8",
    "#abd9e9",
    "#74add1",
    "#4575b4",
]
sample = 10000
fig, ax = plt.subplots(1, 1)
for idx, j in enumerate(hexHtml):
    y = np.random.normal(0, 0.1, size=sample).cumsum()
    x = np.arange(sample)
    ax.scatter(
        x, y, label=str(idx), linewidth=0.2, edgecolors="grey", facecolor=hexHtml[idx]
    )
ax.legend()
plt.show()
```
