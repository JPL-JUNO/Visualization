## 使用颜色参数和颜色映射表

```python
barSlices = 12
theta = np.linspace(0.0, 2 * np.pi, barSlices, endpoint=False)
radii = 30 * np.random.rand(barSlices)
width = 2 * np.pi / barSlices
colors = np.array(
    [
        "c",
        "m",
        "y",
        "b",
        "#C67171",
        "#C1CDCD",
        "#FFEC8B",
        "#A0522D",
        "red",
        "burlywood",
        "chartreuse",
        "green",
    ]
)
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
bars = ax.bar(theta, radii, width=width, color=colors, bottom=0.0)
plt.show()
```

```python
hexHtml = [
    "#d73027",
    "#f46d43",
    "#fdae61",
    "#fee090",
    "#ffffbf",
    "#e0f3f8",
    "#abd9e9",
    "#74add1",
    "#4575b4",
]
sample = 10000
fig, ax = plt.subplots(1, 1)
for idx, j in enumerate(hexHtml):
    y = np.random.normal(0, 0.1, size=sample).cumsum()
    x = np.arange(sample)
    ax.scatter(
        x, y, label=str(idx), linewidth=0.2, edgecolors="grey", facecolor=hexHtml[idx]
    )
ax.legend()
plt.show()
```

散点图的颜色使用模式

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False

a = np.random.randn(100)
b = np.random.randn(100)

exponent = 2

plt.subplot(131)
plt.scatter(
    a,
    b,
    np.sqrt(np.power(a, exponent) + np.power(b, exponent)) * 100,
    c=np.random.randn(100),
    # 参数 c 和关键字参数 cmap 是相互配合使用的一组参数。
    cmap=mpl.cm.jet,
    marker="o",
    zorder=1,
)
plt.subplot(132)
plt.scatter(a, b, 50, marker="o", zorder=10)

plt.subplot(133)
plt.scatter(a, b, 50, c=np.random.rand(100), cmap=mpl.cm.BuPu, marker="+", zorder=100)
plt.show()
```

极区图的颜色使用模式

柱体上色是通过实例方法 `set_facecolor()` 完成的，透明度选择是通过实例方法 `set_alpha()` 实现的。

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

bar_slices = 12
theta = np.linspace(0, 2 * np.pi, bar_slices, endpoint=False)
radii = 30 * np.random.rand(bar_slices)
width = np.pi / 4 * np.random.rand(bar_slices)

fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
bars = ax.bar(theta, radii, width=width, bottom=0.0)
for r, bar in zip(radii, bars):
    bar.set_facecolor(mpl.cm.Accent(r / 30.0))
    bar.set_alpha(r / 30.0)
plt.show()
```

等高线图的颜色使用模式

函数 `contour()` 是用来绘制等高线图的。等高线图是三维图形 z 轴的投影图，既然是投影图就需要刻画三维图形 z 轴的趋势变化，这就是等值线的作用。等值线是将 z 轴上函数值相等的点连接起来的。函数值是通过二元函数计算得出的。不同的等值线需要不同的颜色映射表，等值线也很类似于气象图中的等压线。因此，等高线图也是颜色主题的应用场景之一。

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

s = np.linspace(-0.5, 0.5, 1000)
x, y = np.meshgrid(s, s)

fig, ax = plt.subplots(1, 1)
z = x**2 + y**2 + np.power(x**2 + y**2, 2)
cs = plt.contour(x, y, z, cmap=plt.cm.hot)
plt.clabel(cs, fmt="%3.2f")
plt.colorbar(cs)
plt.show()
```

通过调用函数 `contour()` 获得一个 `ContourSet` 实例，我们可以将 `ContourSet` 实作为参数代入函数 `clabel()`中，为等高线添加标签以此标示出每条等高线的数值大小。我们再将 `ContourSet` 实例传入函数 c`olorbar()` 中为等高线图配置颜色标尺，用来呈现每条等高线的颜色所代表的在 z 轴上的实际投影位置。
