```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False
```

## 共享单一绘图区域的坐标轴

```python
fig, ax1 = plt.subplots()
t = np.arange(0.05, 10, 0.01)
s1 = np.exp(t)
ax1.plot(t, s1, c="b", ls="-")
ax1.set_xlabel("x 坐标轴")
ax1.set_ylabel("以 e 为底指数函数", color="b")
# 将 y 轴标签、主刻度线和刻度标签都设置成蓝色
ax1.tick_params("y", colors="b")

# ax1 shares x-axis with ax2.
ax2 = ax1.twinx()
s2 = np.cos(t**2)
ax2.plot(t, s2, c="r", ls=":")
ax2.set_ylabel("余弦函数", color="r")
ax2.tick_params("y", colors="r")
plt.title("共享单一绘图区域的坐标轴")
plt.show()
```

## 共享不同子区绘图区域的坐标轴

很多时候，我们需要共享不同子区的绘图区域的坐标轴，以求强化绘图区域的展示效果，实现精简绘图区域的目的。这时，我们通过调整函数 `subplots()` 中的参数 `sharey`（或是参数 `sharex`）的不同取值情况，从而实现共享不同子区的绘图区域的坐标轴的需求。

```python
fig, ax = plt.subplots(2, 2)
x1 = np.linspace(0, 2 * np.pi, 400)
y1 = np.cos(x1**2)
x2 = np.linspace(0.01, 10, 100)
y2 = np.sin(x2)
x3 = np.random.rand(100)
y3 = np.linspace(0, 3, 100)
x4 = np.arange(0, 6, 0.5)
y4 = np.power(x4, 3)
ax1 = ax[0, 0]
ax1.plot(x1, y1)
ax2 = ax[0, 1]
ax2.plot(x2, y2)
ax3 = ax[1, 0]
ax3.scatter(x3, y3)
ax4 = ax[1, 1]
ax4.scatter(x4, y4)
plt.suptitle("没有调用参数来作为对比")
plt.show()
```

## 共享个别子区绘图区域的坐标轴

```python
x1 = np.linspace(0, 2 * np.pi, 400)
y1 = np.cos(x1**2)
x2 = np.linspace(0.01, 10, 100)
y2 = np.sin(x2)
x3 = np.random.rand(100)
y3 = np.linspace(0, 3, 100)
x4 = np.arange(0, 6, 0.5)
y4 = np.power(x4, 3)

ax1 = plt.subplot(221)
ax1.plot(x1, y1)
ax2 = plt.subplot(222)
ax2.plot(x2, y2)
ax3 = plt.subplot(223)
ax3.scatter(x3, y3)
ax4 = plt.subplot(224, sharex=ax1)
ax4.scatter(x4, y4)
plt.suptitle("用函数 autoscale()调整坐标轴范围")
plt.show()
```

### 延伸阅读——用函数 autoscale()调整坐标轴范围

如果我们对某一个子区的坐标轴范围和数据范围的搭配比例不是很满意，可以使用函数 `autoscale()` 进行坐标轴范围的自适应调整，以使图形可以非常紧凑地填充绘图区域。调用签名是 `autoscale(enable=True,axis="both",tight=True)`。调用签名中的具体参数的含义如下所示。
- enable：进行坐标轴范围的自适应调整。
- axis：使 x、y 轴都进行自适应调整。
- tight：让坐标轴的范围调整到数据的范围上。

```python
x1 = np.linspace(0, 2 * np.pi, 400)
y1 = np.cos(x1**2)
x2 = np.linspace(0.01, 10, 100)
y2 = np.sin(x2)
x3 = np.random.rand(100)
y3 = np.linspace(0, 3, 100)
x4 = np.arange(0, 6, 0.5)
y4 = np.power(x4, 3)

ax1 = plt.subplot(221)
ax1.plot(x1, y1)
ax2 = plt.subplot(222)
ax2.plot(x2, y2)
ax3 = plt.subplot(223)
ax3.scatter(x3, y3)
plt.autoscale(enable=True, axis="both", tight=True)
ax4 = plt.subplot(224, sharex=ax1)
ax4.scatter(x4, y4)
plt.suptitle("用函数 autoscale()调整坐标轴范围")
plt.show()
```
