## 保存输出图形

在我们执行完代码后，通常会有图形输出界面，此时可以有两种保存图形的方法：一种是在图形输出界面单击“保存”按钮；一种是通过执行代码进行输出图形的保存。

### 13.2.1 方法 1——使用“保存”按钮进行存储

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

fig, ax1 = plt.subplots()
t = np.arange(0.05, 10, 0.01)
s1 = np.exp(t)
ax1.plot(t, s1, c="b", ls="-")
ax1.set_xlabel("x axis")
ax1.set_ylabel("base e", color="b")
ax1.tick_params("y", colors="b")
ax2 = ax1.twinx()
s2 = np.cos(t**2)
ax2.plot(t, s2, c="r", ls=":")
ax2.set_ylabel("cos", color="r")
ax2.tick_params("y", colors="r")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

fig, ax1 = plt.subplots()
t = np.arange(0.05, 10, 0.01)
s1 = np.exp(t)
ax1.plot(t, s1, c="b", ls="-")
ax1.set_xlabel("x axis")
ax1.set_ylabel("base e", color="b")
ax1.tick_params("y", colors="b")
ax2 = ax1.twinx()
s2 = np.cos(t**2)
ax2.plot(t, s2, c="r", ls=":")
ax2.set_ylabel("cos", color="r")
ax2.tick_params("y", colors="r")
plt.savefig("./FIGURE_DEMO.png")
plt.show()
```
