```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0.1, 0.6, 10)
y = np.exp(x)
error = 0.05 + 0.15 * x
lower_error = error
upper_error = 0.3 * error
# value - lower_error, value + upper_error
error_limit = (lower_error, upper_error)
plt.errorbar(x, y, yerr=error_limit, fmt=':o', ecolor='y',
             elinewidth=4, ms=5, mfc='c', mec='r', capthick=1, capsize=2)
# mec: marker edge color
# ms: marker size
# mfc: marker face size
plt.xlim(0, 0.7)
plt.show()
```

我们采用单一数值的非对称形式的误差棒，函数 errorbar()的参数含义如下所示。
- x，y：数据点的位置。
- yerr：单一数值的非对称形式误差范围。
- fmt：数据点的标记样式和数据点标记的连接线样式。
- ecolor：误差棒的线条颜色。
- elinewidth：误差棒的线条粗细。
- ms：数据点的大小。
- mfc：数据点的标记颜色。
- mec：数据点的标记边缘颜色。
- capthick：误差棒边界横杠的厚度。
- capsize：误差棒边界横杠的大小。

### 案例 1——带误差棒的柱状图

```python
x = np.arange(5)
y = [100, 68, 79, 91, 82]
std_err = [7, 2, 6, 10, 5]

error_attri = dict(elinewidth=2, ecolor='black', capsize=3)

plt.bar(x, y, color='c', width=0.6, align='center', yerr=std_err,
        error_kw=error_attri)
plt.grid(True, axis='y', ls=':', color='gray', alpha=.2)
plt.show()
```

### 案例 2——带误差棒的条形图

```python
x = np.arange(5)
y = [1200, 2400, 1800, 2200, 1600]
std_err = [150, 100, 180, 130, 80]

bar_width = 0.6
colors = ["#e41a1c", "#377eb8", "#4daf4a", "#984ea3", "#ff7f00"]
plt.barh(x, y, bar_width, color=colors, align="center", xerr=std_err)
plt.grid(True, axis="x", ls=":", color="gray", alpha=0.2)
plt.xlim(0, 2600)
plt.show()
```

带误差棒的条形图的绘制是通过使用函数 `barh()` 中的关键字参数 `xerr` 实现的。

```python
x = np.arange(5)
y1 = [100, 68, 79, 91, 82]
y2 = [120, 75, 70, 78, 85]
std_err1 = [7, 2, 6, 10, 5]
std_err2 = [5, 1, 4, 8, 9]

error_attri = dict(elinewidth=2, ecolor="black", capsize=3)
bar_width = 0.4
plt.bar(
    x,
    y1,
    bar_width,
    color="#87CEEB",
    align="center",
    yerr=std_err1,
    error_kw=error_attri,
    label="2010",
)
plt.bar(
    x + bar_width,
    y2,
    bar_width,
    color="#CD5C5C",
    align="center",
    yerr=std_err2,
    error_kw=error_attri,
    label="2013",
)
plt.xticks(x + bar_width / 2, list("ABCDE"))
plt.grid(True, axis="y", ls=":", color="gray", alpha=0.2)
plt.legend()
plt.show()
```

### 案例 4——带误差棒的堆积柱状图

```python
x = np.arange(5)
y1 = [100, 68, 79, 91, 82]
y2 = [120, 75, 70, 78, 85]
std_err1 = [7, 2, 6, 10, 5]
std_err2 = [5, 1, 4, 8, 9]

error_attri = dict(elinewidth=2, ecolor="black", capsize=3)
plt.bar(
    x,
    y1,
    bar_width,
    color="#6495ED",
    align="center",
    yerr=std_err1,
    error_kw=error_attri,
    label="2010",
)
plt.bar(
    x,
    y2,
    bar_width,
    bottom=y1,
    color="#CD5C5C",
    align="center",
    yerr=std_err2,
    error_kw=error_attri,
    label="2013",
)
plt.xticks(x, list("ABCDE"))
plt.grid(True, axis="y", ls=":", color="gray", alpha=0.2)
plt.legend()
plt.show()
```
